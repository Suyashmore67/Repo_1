```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
Population_data = pd.read_csv(r"C:\Users\suyas\OneDrive\Documents\Desktop\task 1\world_population.csv")
```

```python
# Display basic information about the dataset
print("Datasert Preview:")
print(Population_data.head())
print("\nColumns in the dataset:", Population_data.columns)
```

```python
# Step 1: Filter the data for a specific year
filtered_data = Population_data[['Country/Territory', '2022 Population']].dropna()
filtered_data.columns = ['Country', 'Population']
```

```python
# Step 2: Sort by population and filter for the top 10 countries
top10_countries = filtered_data.sort_values(by='Population', ascending=False).head(10)
```

```python
# Step 3: Plot a bar chart for the Top 10 Most Populated Countries
plt.figure(figsize=(12, 6))
sns.barplot(x='Population', y='Country', data=top10_countries, palette='viridis')
plt.title('Top 10 Most Populated Countries in 2022', fontsize=16)
plt.xlabel('Population', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
# Step 4: Plot a histogram for population distribution
plt.figure(figsize=(12, 6))
sns.histplot(filtered_data['Population'], bins=30, kde=True, color='blue', edgecolor='black')
plt.title('Population Distribution Across All Countries in 2022', fontsize=16)
plt.xlabel('Population', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```
