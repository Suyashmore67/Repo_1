```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
# Load the uploaded CSV file to understand its structure
file_path = r"C:\Users\suyas\OneDrive\Documents\Desktop\task 4\twitter_training.csv"
data = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
data.head(), data.info()
```

```python
# Renaming columns for clarity
data.columns = ["ID", "Topic", "Sentiment", "Text"]

# Checking for missing values and dropping rows if necessary
data = data.dropna(subset=["Text"])

# Displaying the cleaned data structure
data.head(), data["Sentiment"].value_counts()
```

```python
# Setting the style for plots
sns.set(style="whitegrid")

# Visualizing sentiment distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x="Sentiment", palette="viridis", order=["Positive", "Negative", "Neutral", "Irrelevant"])
plt.title("Sentiment Distribution in Social Media Data", fontsize=16)
plt.xlabel("Sentiment", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.show()

# Generating a word cloud for positive sentiment
positive_text = " ".join(data[data["Sentiment"] == "Positive"]["Text"])
wordcloud = WordCloud(width=800, height=400, background_color="white", colormap="viridis").generate(positive_text)

# Displaying the word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud for Positive Sentiments", fontsize=16)
plt.show()
```
